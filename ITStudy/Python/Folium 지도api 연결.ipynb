```python
#페리카나 매장 데이터 지도에 마커 찍기

import sys
import urllib.request
import datetime
import time
import json
import folium
import pandas as pd

df = pd.read_csv( '../data/pericana.csv', thousands=',', encoding='euc-kr')
pericana_loc = df.store_address
data_address= []
for address in pericana_loc:
        data_address.append(address)
data_address

def get_request_url(url) :
    
    req = urllib.request.Request(url)
    req.add_header("X-Naver-Client-id", '8Ns_1Ml4EJEnZPpWkjfP')
    req.add_header("X-Naver-Client-Secret",'EY7gtN9qnP')
    try :
        response = urllib.request.urlopen(req)
        if response.getcode() == 200:
            print ("[%s] Url Request Success" % datetime.datetime.now())
            return response.read().decode('utf-8')
    except Exception as e:
        print(e)
        print("[%s] Error for URL : %s" % (datetime.datetime.now(), url))
        return None
    
def getGeoData(address):
        base = "https://openapi.naver.com/v1/map/geocode"
        node = ""
        try :
            parameters = "?query=%s" % urllib.parse.quote(address)
        except Exception as e :
            return None
        url = base + node + parameters
    
        retData = get_request_url(url)
    
        if (retData == None) :
            return None
        else : 
            return json.loads(retData)
    

map_osm = folium.Map(location=[37.5502, 126.982], zoom_start=8)

for i in range (0,10) :
    jsonResult = getGeoData(data_address[i])
    if jsonResult ==None :
        continue
   
    if 'result' in jsonResult.keys():
        print('검색어', jsonResult['result']['userquery'])
        
        for item in jsonResult['result']['items']:
            print('==================')
            print('주소 : ', item['address'])
            print('위도 : ', str(item['point']['y']))
            print('경도 : ', str(item['point']['x']))
            
            folium.Marker([item['point']['y'], item['point']['x']]).add_to(map_osm)

map_osm
```
